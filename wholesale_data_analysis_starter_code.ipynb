## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']

# Display the first few rows of the DataFrame to verify the new column
print(df[['unit_price', 'qty', 'subtotal']].head())
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
# Create shipping_price column
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

# Create subtotal column
df['subtotal'] = df['unit_price'] * df['qty']

# Display the first few rows to verify the new columns
print(df[['unit_weight', 'qty', 'total_weight', 'shipping_price', 'unit_price', 'subtotal']].head())
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Define the sales tax rate
SALES_TAX_RATE = 0.0925  # 9.25%

# Calculate the total price
df['total_price'] = (df['subtotal'] + df['shipping_price']) * (1 + SALES_TAX_RATE)

# Display the first few rows to verify the new column
print(df[['subtotal', 'shipping_price', 'total_price']].head())

```

```python
# Create a column for the cost

# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
# Calculate the cost for each line
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# Display the first few rows to verify the new column
print(df[['unit_cost', 'qty', 'shipping_price', 'line_cost']].head())
```

```python
# Create a column for the profit of
# each line using line cost and line price
# Calculate the profit for each line
df['line_profit'] = (df['subtotal'] + df['shipping_price']) - df['line_cost']

# Display the first few rows to verify the new column
print(df[['subtotal', 'shipping_price', 'line_cost', 'line_profit']].head())
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# Step 1: Identify the top 5 clients by quantity
top_5_clients = df.groupby('client_id')['qty'].sum().nlargest(5)

print("Top 5 clients by quantity:")
print(top_5_clients)
print()

# Step 2: Calculate total spending for each of these clients
for client_id in top_5_clients.index:
    client_spending = df[df['client_id'] == client_id]['total_price'].sum()
    print(f"Client ID {client_id} spent: ${client_spending:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Identify the top 5 clients by quantity
top_5_clients = df.groupby('client_id')['qty'].sum().nlargest(5).index

# Create the summary DataFrame
summary = df[df['client_id'].isin(top_5_clients)].groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'line_profit': 'sum'
})

# Rename columns for clarity
summary.columns = ['Total Units', 'Total Shipping', 'Total Revenue', 'Total Profit']

# Sort by Total Profit
summary_sorted = summary.sort_values('Total Profit', ascending=False)

# Display the summary
print(summary_sorted)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Convert currency columns to millions of dollars
currency_columns = ['Total Revenue ($ millions)', 'Total Profit ($ millions)']
df[currency_columns] = df[currency_columns] / 1e6

# Convert Total Shipping Cost to dollars (it's already in dollars)
df['Total Shipping Cost ($)'] = df['Total Shipping Cost ($)'].round(2)

# Sort the DataFrame by 'Total Profit ($ millions)' in descending order
df_sorted = df.sort_values(by='Total Profit ($ millions)', ascending=False)

# Round all currency columns to 2 decimal places
df_sorted[currency_columns + ['Total Shipping Cost ($)']] = df_sorted[currency_columns + ['Total Shipping Cost ($)']].round(2)

# Display the sorted and formatted DataFrame
print(df_sorted)
```
